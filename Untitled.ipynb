```python
import pandas as pd

# CSV dosyasının adını doğru olarak kullanın
data = pd.read_csv("tmdb-movie-metadata/tmdb_5000_movies.csv")  # Dosya adını kontrol edin
print(data.head())  # İlk birkaç satırı göster
```

```python
import pandas as pd
from sklearn.preprocessing import MultiLabelBinarizer

# Örnek: Büyük veri setini temsil eden DataFrame
# title, vote_average, vote_count ve kodlanmis_tür sütunlarına sahip
# 'df' adında bir DataFrame olduğunu varsayıyoruz

# Kodlanmış türler sütununda One-Hot Encoding uygulama
mlb = MultiLabelBinarizer()
one_hot_encoded = mlb.fit_transform(data['kodlanmis_tür'])

# Tür kodlarına dayalı sütun isimleri oluşturma
column_names = [f"Tür_{code}" for code in mlb.classes_]

# Yeni sütunları DataFrame'e ekleme
one_hot_df = pd.DataFrame(one_hot_encoded, columns=column_names, index=data.index)
data = pd.concat([data, one_hot_df], axis=1)

# 'kodlanmis_tür' sütununu kaldırmak isteyebilirsiniz (isteğe bağlı)
data.drop(columns=['kodlanmis_tür'], inplace=True)

# Sonuçları kaydetme (isteğe bağlı)
data.to_csv("cleaned_movies_dataset.csv", index=False)

print("Güncellenmiş veri seti başarıyla işlendi ve kaydedildi.")
```

```python
# Tür bilgilerini birleştirin
def create_genre_features(data):
    genre_cols = [col for col in data.columns if col.startswith('Tür_')]
    data['genres_combined'] = data[genre_cols].apply(lambda x: ' '.join([col.replace('Tür_', '') for col, val in x.items() if val == 1]), axis=1)
    return data

# Kalite skoru oluşturun
data['quality_score'] = data['vote_average'] * np.log1p(data['vote_count'])
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# TF-IDF ile tür benzerliği
tfidf = TfidfVectorizer()
tfidf_matrix = tfidf.fit_transform(data['genres_combined'])

# Cosine similarity hesaplama
cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)
```

```python
def create_genre_features_fixed(df):
    # Tür kolonlarını bul
    genre_cols = [col for col in df.columns if col.startswith('Tür_')]
    
    print(f"Bulunan {len(genre_cols)} tür kolonu: {genre_cols[:5]}...")  # İlk 5'ini göster
    
    if not genre_cols:
        print("HATA: Tür kolonları bulunamadı!")
        return df
    
    # Her satır için aktif türleri birleştir
    def combine_genres(row):
        active_genres = []
        for col in genre_cols:
            if pd.notna(row[col]) and row[col] == 1:  # NaN kontrolü ekleyelim
                # 'Tür_' prefix'ini kaldır ve genre name'i al
                genre_name = col.replace('Tür_', f'Genre{col.split("_")[1]}')
                active_genres.append(genre_name)
        return ' '.join(active_genres) if active_genres else 'Unknown'
    
    # Yeni kolonu oluştur
    df = df.copy()  # Orijinal veriyi değiştirmeyelim
    df['genres_combined'] = df.apply(combine_genres, axis=1)
    
    return df

# Fonksiyonu uygula
data_with_genres = create_genre_features_fixed(data)

# Sonucu kontrol et
print("\ngenres_combined oluşturuldu mu?")
print('genres_combined' in data_with_genres.columns)

print("\nİlk birkaç genres_combined değeri:")
print(data_with_genres[['title', 'genres_combined']].head())
```

```python
# Eğer genres_combined başarıyla oluşturulduysa
if 'genres_combined' in data_with_genres.columns:
    print("genres_combined kolonu bulundu, devam ediyoruz...")
    
    from sklearn.feature_extraction.text import TfidfVectorizer
    from sklearn.metrics.pairwise import cosine_similarity
    
    # Boş değerleri kontrol et
    print(f"Boş genres_combined değer sayısı: {data_with_genres['genres_combined'].isna().sum()}")
    print(f"'Unknown' değer sayısı: {(data_with_genres['genres_combined'] == 'Unknown').sum()}")
    
    # TF-IDF uygula
    tfidf = TfidfVectorizer(stop_words='english')
    tfidf_matrix = tfidf.fit_transform(data_with_genres['genres_combined'])
    
    print(f"TF-IDF matrix boyutu: {tfidf_matrix.shape}")
    
    # Cosine similarity hesapla
    cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)
    print(f"Cosine similarity matrix boyutu: {cosine_sim.shape}")
    
    print("Başarılı! Artık öneri sistemi hazır.")
else:
    print("HATA: genres_combined kolonu oluşturulamadı!")
```

```python
def get_recommendations(title, cosine_sim=cosine_sim, df=data_with_genres, top_n=10):
    """
    Verilen film başlığına göre benzer filmler önerir
    """
    # Film var mı kontrol et
    if title not in df['title'].values:
        print(f"'{title}' filmi bulunamadı!")
        print("Mevcut filmlerden bazıları:")
        print(df['title'].head(10).tolist())
        return None
    
    # Film indeksini bul
    idx = df[df['title'] == title].index[0]
    
    # Benzerlik skorlarını al
    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    
    # En benzer filmleri al (kendisi hariç)
    sim_scores = sim_scores[1:top_n+1]
    movie_indices = [i[0] for i in sim_scores]
    
    # Sonuçları döndür
    recommendations = df.iloc[movie_indices][['title', 'vote_average', 'vote_count', 'genres_combined']].copy()
    recommendations['similarity_score'] = [score[1] for score in sim_scores]
    
    return recommendations.round({'vote_average': 1, 'similarity_score': 3})

# Test edelim
print("=== FİLM ÖNERİ SİSTEMİ TESTİ ===")

# Avatar'ı test edelim
if 'Avatar' in data_with_genres['title'].values:
    print("\nAvatar için öneriler:")
    avatar_recs = get_recommendations('Avatar')
    print(avatar_recs)
else:
    # İlk filmi test edelim
    test_movie = data_with_genres['title'].iloc[0]
    print(f"\nTest filmi: {test_movie}")
    recommendations = get_recommendations(test_movie)
    print(recommendations)

# Sistem istatistikleri
print("\n=== SİSTEM İSTATİSTİKLERİ ===")
print(f"Toplam film sayısı: {len(data_with_genres)}")
print(f"'Unknown' türü olan filmler: 28")
print(f"Benzersiz tür kombinasyonu: {data_with_genres['genres_combined'].nunique()}")
```
