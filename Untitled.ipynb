```python
import pandas as pd

# CSV dosyasının adını doğru olarak kullanın
data = pd.read_csv("tmdb-movie-metadata/tmdb_5000_movies.csv")  # Dosya adını kontrol edin
print(data.head())  # İlk birkaç satırı göster
```

```python
import pandas as pd
import numpy as np
data=data.drop("tagline", axis=1)
data.head()
```

```python
import json
data['genres'] = data['genres'].apply(json.loads)
data['tür'] = data['genres'].apply(lambda x: [d['name'] for d in x])

print(data)
```

```python
print(data["tür"])
```

```python
data=data.drop(["homepage","budget","original_language","original_title","overview",	"popularity","production_companies","production_countries"],axis=1)
```

```python
data
```

```python
data=data.drop(["genres","id","keywords","revenue","spoken_languages","status"],axis=1)
```

```python
data.head()
```

```python
benzersiz_tür=set(genre for sublist in data["tür"]for genre in sublist)
benzersiz_tür=list(benzersiz_tür)
print("benzersiz türler:")
print(benzersiz_tür)
```

```python
benzersiz_tür_indexi={genre : idx for idx, genre in enumerate(benzersiz_tür)}
print("Türlere karşılık kodlar:")
print(benzersiz_tür_indexi)
```

```python
kodlanmis_türler = data["tür"].apply(lambda genres: [benzersiz_tür_indexi[genre] for genre in genres])
data["kodlanmis_tür"]=kodlanmis_türler
print(data)
```

```python
data=data.drop(["release_date","runtime"],axis=1)
```

```python
data.head()
```

```python
import pandas as pd
from sklearn.preprocessing import MultiLabelBinarizer

# Örnek: Büyük veri setini temsil eden DataFrame
# title, vote_average, vote_count ve kodlanmis_tür sütunlarına sahip
# 'df' adında bir DataFrame olduğunu varsayıyoruz

# Kodlanmış türler sütununda One-Hot Encoding uygulama
mlb = MultiLabelBinarizer()
one_hot_encoded = mlb.fit_transform(data['kodlanmis_tür'])

# Tür kodlarına dayalı sütun isimleri oluşturma
column_names = [f"Tür_{code}" for code in mlb.classes_]

# Yeni sütunları DataFrame'e ekleme
one_hot_df = pd.DataFrame(one_hot_encoded, columns=column_names, index=data.index)
data = pd.concat([data, one_hot_df], axis=1)

# 'kodlanmis_tür' sütununu kaldırmak isteyebilirsiniz (isteğe bağlı)
data.drop(columns=['kodlanmis_tür'], inplace=True)

# Sonuçları kaydetme (isteğe bağlı)
data.to_csv("cleaned_movies_dataset.csv", index=False)

print("Güncellenmiş veri seti başarıyla işlendi ve kaydedildi.")
```
